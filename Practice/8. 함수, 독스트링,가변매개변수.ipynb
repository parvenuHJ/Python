## 함수(function)
- 함수란?
    - 호출 될 때 틀정 기능을 실행하는 재사용 가능한 코드블록
- 함수의 목적
    - 반복적인 프로그래밍을 피할 수 있다.
    - 모듈화(코드모음)으로 인해 전체적인 코드의 가독성이 좋아진다.
    - 프로그램 문제가 발생하거나 기능의 변경이 필요할 때, 유지보수가 용이하다.

```python
# 함수 정의
def number_sum(num1, num2) :
    result = num1 + num2
    return result
```

```python
n1 = 3
n2 = 10
print(number_sum(n1,n2))
```

```python
# 두 수를 입력 받아서 뺀 결과를 return하는 함수를 정의
num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))

def number_sub(num1, num2) :
    if num1 >= num2 :
        result = num1 - num2
    else :
        result = num2 - num1
    return result


print(number_sub(num1, num2))

        
```

```python
# 두 수를 입력 받아서 원하는 연산을 수행하여 결과를 return 하는 함수를 정의
num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))
op = input("연산자 입력(+,-) >> ")

def num_cal(n1, n2, op) :
    if op == '+' :
        result = num1 + num2
    elif op == '-' :
        # 삼항연산자 이용
        result = num1-num2 if n1 > n2 else num2 - num1
    else :
        result = '연산자를 다시 입력'
    return result

print("결과 : " , num_cal(num1,num2,op))
```

### 독스트링(docstring)
- 함수의 설명을 작성 → Shift + Tab 으로 확인 가능
- 독스트링 정의 하는 방법 : """설명문"""

```python
num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))
op = input("연산자 입력(+,-) >> ")

def num_cal(n1, n2, op) :
    """두 수를 입력받아서 원하는 연산을 수행하는 함수"""# 독스트링
    if op == '+' :
        result = num1 + num2
    elif op == '-' :
        # 삼항연산자 이용
        result = num1-num2 if n1 > n2 else num2 - num1
    else :
        result = '연산자를 다시 입력'
    return result

print("결과 : " , num_cal(num1,num2,op))
```

### 함수 매개변수 - 가변 매개변수(*)
- 함수 호출 시 몇 개의 인수가 전달될지 알 수 없다면, 사용자가 직접 매개변수의 개수를 정할 수 있도록 선언

```python
# 가변 매개변수로 함수를 정의
def add(*args) :
    result = sum(args)
    return result
```

```python
add(1,2,3,4,5)
```


- 순서 주의 : (*args , a) -> *args는 매개변수가 무한정으로 받고있어서 a를 받을 수 가 없음
