```python
# 리스트는 가변형이므로 주소값 공유

list1 = [1, 2, 3]
list2 = list1
list2.append(4)

print(f"list1: {list1}, list2: {list2}")
print(f"list1 id: {id(list1)}, list2 id: {id(list2)}")

# 숫자는 불변형이므로 주소값 변경
# num1 , num2 는 이미 hip에 한자리를 차지하고있음(불변)
num1 = 10
num2 = num1
num2 += 5

print(f"num1: {num1}, num2: {num2}")
print(f"num1 id: {id(num1)}, num2 id: {id(num2)}")
```

- 레퍼런스형 대입 ->  Stack 영역(id값 = 주소값) -> Hip 영역(값 출력)
- 일반형 대입 ->  Hip 영역(값 출력)
- ex)
- " list1(레퍼런스형) = ~~~, "
- "list2 = list1"               =>  [ list1 -> stack 영역(id값)-> list2 ] 이여서
- list1의 id값 = list2의 id값
-
-
- 그래서 바로 위의 예시에서 
- 주소값 : list1 = list2
- 주소값 : num1 != num2

```python
# 주소값이 list1 과 list2가 같은데, 다르게 설정해주고 싶으면
list1 = [1, 2, 3]
list2 = list1.copy()
list2.append(4)

print(f"list1: {list1}, list2: {list2}")
print(f"list1 id: {id(list1)}, list2 id: {id(list2)}")
```

### 딕셔너리
- 말 그대로 사전이라는 뜻
- Key는 불변적, Value는 가변적
- 딕셔너리명 = {'key' : value, 'key' : value ... }


```python
# 딕셔너리 생성
dic1 = {"key":"Value","name":"김훈종", "age" : 28, "phone" : "010-1234-1234"} 
dic1
```

```python
# 딕셔너리 값 추가 → 딕셔너리는 순서가 없음
# dic['key'] = value
print(dic1['key']) # 딕셔너리 인덱싱
dic1['birth'] = '01/01'
dic1
```

```python
# 딕셔너리 값 삭제
# del 딕셔너리명['key']
del dic1['age']
dic1
```

```python
# 딕셔너리 값 가져오기 1 → 인덱싱
dic1['name']
```

```python
# 딕셔너리 값 가져오기 2 → get()
dic1.get('name')
```

```python
print(dic1['name'])
```

```python
print(dic1.get('name'))
```

#### 딕셔너리 값 가져오기 1 → 인덱싱 , 2 → get() 의 차이점
- 인덱싱 : 오류가 남
- get() : 오류가 안나고, None 으로 출력됨

```python
print(dic1['age'])
```

```python
print(dic1.get('age'))
```

```python
# key 값만 가져오는 함수 → keys() 함수
dic1.keys()
```

#### .keys() 는 인덱스가 없어서 순서가 없음 , dict_keys , dict_vaules 라는 자료형임
#### list( 딕셔너리명.keys()) 로 list 로 형변환 하게되면 가능

```python
list(dic1.keys())
```

```python
# value 값만 가져오는 함수 → values() 함수
dic1.values()
```

```python
list(dic1.values())
```

#### 딕셔너리는 순서가 없으나(시퀀스가 아니지만), for 문의 활용이 가능하다!
- for ~ in 시퀀스 :

```python
for key in dic1.keys():
    print(key)
```

```python
# items() → (key, value)
for items in dic1.items():
    print(items)
```

```python
# 변수명 2개를 주면 알아서 key,value 값이 출력됨
for a, b in dic1.items():
    print(a,b )
```
