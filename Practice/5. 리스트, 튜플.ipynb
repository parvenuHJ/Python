### 리스트란?
- 순서가 있고, 수정 가능한 객체의 집합 : 수정 삭제, 추가 기능
- 대괄호로 작성하고, 내부 요소는 콤마(,)로 구분

#### 리스트 인덱싱
- 리스트명[인덱스번호]

```python
list2 = [1,2,3,['a','b','c']]
temp = list2[3]
print(temp)
print(temp[1])

# 2차원 인덱싱
print(list2[3][1])
```

#### 리스트 슬라이싱
- 리스트명[start : end]
- 시작값은 포함, 끝값은 미포함

```python
list3 = [0,1,2,3,4]
# 1,2 요소에 접근하기 위해서는
list3[1:3]
```

- list와 list를 더하면?

```python
# list와 list를 더하면. 뒤로 붙게 된다.
# ※ 단, 숫자 연산은 안된다!
list4 = [1,2,3]
list5 = [3,4,5,6]
list4 + list5
```

###  리스트 관련함수
#### 리스트 값 추가 : append(), insert()
- 리스트.append(값) : 리스트의 맨 뒤에 값을 추가
- 리스트.insert(인덱스위치, 값) : 사용자가 지정한 위치에 값을 추가
#### 리스트 값 수정 
- 리스트.[인덱스번호] = 수정할 값 : 해당 위치에 값을 수정하겠다.(값이 교체됨)
#### 리스트 값 삭제 : remove(), del 객체
- 리스트.remove(값) : 해당하는 값을 삭제
- del 리스트[인덱스번호] : 해당하는 위치의 값을 삭제
- del 리스트[] : 해당하는 리스트를 삭제


```python
# 리스트 값 추가
list_fruit=['사과','배','포도']
list_fruit.append('복숭아')
list_fruit.insert(2,'딸기')
list_fruit
```

```python
# 리스트 값 수정
list_fruit[0] = '수박'
list_fruit
```

```python
#리스트 값 삭제 remove()함수 사용
list_fruit.remove('배')
list_fruit
```

```python
# 리스트 값 삭제 del 객체 사용
del list_fruit[3]
list_fruit
```

```python
# 주로 사용하지는 않음 -> pop() : 해당하는 위치의 값을 삭제하는데, 남긴다.
strawBerry = list_fruit.pop(1)
list_fruit
```

```python
strawBerry
```

```python
waterMelon = list.fruit.remove('수박')
waterMelon
```

### 리스트 정렬 : sort() ,  reverse()
- 리스트.sort() : 오름차순으로 정렬
- 리스트.reverse() : 역순으로 요소들을 뒤집기
    - '내림차순'을 하고 싶은 경우 : 리스트.sort() -> 리스트.reverse() 을 연속으로 사용

```python
list_nums = [9,11,111,12,1,5,6]
list_nums.sort()
list_nums
```

```python
list_nums = [9,11,111,12,1,5,6]
list_nums.reverse()
list_nums
```

```python
# 내림차순
list_nums.sort()
list_nums.reverse()
list_nums
```

### 리스트 길이 확인
- len(리스트) : 리스트의 길이를 반환

```python
len(list_nums)
```

## 튜플(tuple)
- 파이썬의 자료형 중 하나
- 순서(인덱스)가 있는 객체의 집합
- 소괄호로 작성되어지며, 튜플의 내부 요소들은 콤마(,)로 구분
#### 튜플은 추가,수정,삭제 불가능!!

```python
tuple1 = (0,1,2,3,('a','b','b'),5)
tuple1
```

```python
tuple1[3]
```

```python
tuple1[4][1]
```

```python
tuple1[0:2]
```

```python
tuple1[4:5]
```

```python
tuple1[4:6]
```

```python
# 튜플을 2번째 인덱스 부터 슬라이싱
print(tuple1[2:])
```

```python
#튜플 길이
len(tuple1)
```

```python
#튜플은 추가, 수정, 삭제가 불가능하다!!
tuple1[1] = "B"
print(tuple1)
```

```python
#튜플은 추가, 수정, 삭제가 불가능하다!!
tuple1.insert(3,"B")
print(tuple1)
```

```python
#tuple을 list 형태로 형변환은 가능
# () -> [] 
#tuple -> list
list(tuple1)
```

### 리스트와 튜플의 공통점과 차이점
- 공통점
    - 타입과 상관없이 일련의 요소들을 가질 수 있음
    - [1,"A",[2,"B"]]   ,  (1,"A",(2,"B")) 가능
    - 요소의 순서(인덱스)를 관리한다. -> 인덱싱과 슬라이싱이 가능!
- 차이점
    -리스트 []
        - 가변적(추가, 수정, 삭제 가능)
        - 요소가 몇 개 들어갈지 명확하지 않는 경우에 사용
    - 튜플 ()
        - 불변적(추가, 수정, 삭제 불가능)
        - 요소 개수가 사전에 명확히 알고 있을 경우에 사용

```python
# 튜플 안의 리스트는 변경가능!!!
tuple2 = (0,1,2,3,['a','b','c'],5)
tuple2[4][1] = 'd'
tuple2
```

```python
# 튜플의 요소를 바꾸는 것은 불가능!!
tuple2[4] = ['e']
```
