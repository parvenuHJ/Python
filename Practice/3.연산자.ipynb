### 연산자의 종류

1. 산술 연산자 +   -   *   /   //   %
2. 지수 연산자 **
3. 대입(복합)연산자  =   +=   -=   *=   /=   //=   %=
4. 관계(비교) 연산자 : 두 수를 비교하여 참 또는 거짓을 반환.  >    >=   <   <=   ==   !=
5. 논리 연산자 : 조건문을 연결하기 위해 사용.  and or not
6. 멤버 연산자 : 해당 값이 대상 객체에 존재하는지 확인하기 위해 사용.  in   not in
6. 삼항 연산자 a if 조건식 else b

###  alt 후 커서 쭉 내리면 한꺼번에 출력가능!

```python
# 산술연산자
# 나누기(/) ,나머지(%), 몫(//)

num1 = 10
num2 = 7

print(f"나누기 : {num1 / num2 }")
# :.2f -> 소수점 2째짜리까지만 출력, 알아서 반올림
# 포매팅에서만 사용가능!!
print(f"나누기 : {num1 / num2 : .2f}")
print(f"나머지 : {num1 % num2}")
print(f"몫 : {num1 // num2}")
```

### 파이썬에서는 boolean의 True, False를 쓸 때 유의! 첫번째 알파벳을 대문자로

- 랜덤함수

```python
import random as rd
```

```python
#randint(시작값,끝값) -> 랜덤함수는 끝값을 포함해서 나옴
a = rd.randint(1,10)
b = rd.randint(1,10)
print(f"a:{a},b:{b}")

print(a < b)
print(a > b)
print(a == b)
print(a != b)
```

### 논리 연산자
- True, False의 논리값(boolean값)을 반환하는 연산자
- not 연산자

```python
a = 3
b = 7
print(f'a < b :{a < b}')
print(f'a < b :{not a < b}')
```

- and 연산자
    - 두 조건이 모두 참(True)일 경우 참(True)으로 반환

```python
print(3 < 5) 
print(10 < 20)

3 < 5 and 10 < 20
print( 3 < 5 and 10 < 20 )
```

- or 연산자
    - 두 조건이 하나라도 참(True)이면 참 반환

```python
print(3 > 5) 
print(10 < 20)

print( 3 > 5 or 10 < 20 )
```

### 삼항 연산자
    - a if 조건식 else b
    - a(참일 때 실행) if 조건식 else b(거짓일 때 실행)

```python
score = 80
"합격" if score >= 60 else "불합격"
# 조건식(score >= 60 )이 True-> "합격", False->"불합격"
```
