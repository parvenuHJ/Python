### 문자열 인덱싱과 슬라이싱
- 인덱싱 : 요소 하나를 가리키는 것 -> 하나하나를 "가리킨다"
- 슬라이싱 : 요소들을 잘라오는 것  -> 범위를 지정해서, 어디부터 어디까지를 "잘라낸다"
    - s[start : end] = start는 포함, end는 미포함!!!!

- 인덱스 번호는 앞에서 0부터, 뒤로 세면 -1 부터 시작

###  문자열 포맷코드
- 문자열 내부에 값 삽입


```python
#문자열 포매팅(Formatting)
#문자열 안의 특정한 값을 삽입해야 될때 사용

# 1. % 기호사용
day = 24
month =7

# " " 혹은 ' ' 둘다 문자하나 나 문자열이나 모두 문자열로 인식하기 때문에 같음
s = '오늘은 %d월 %d일 입니다.'%(month, day)
s
```

###  Tab -> 자동완성

```python
# 2. format 함수 사용
day = 24
month =7

s = "오늘은 {}월 {}일 입니다." .format(month, day)
s
```

```python
# 3. f 문자열 사용
day = 24
month = 7

s = f"오늘은 {month}월 {day}일 입니다."
s
```
