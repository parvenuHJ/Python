```python
#         return   
num1 = int(input('    >> '))
num2 = int(input('    >> '))
op = input(" (+,-) >> ")

def num_cal(n1, n2, op) :
    if op == '+' :
        result = num1 + num2
    elif op == '-' :
        #  
        result = num1-num2 if n1 > n2 else num2 - num1
    else :
        result = '  '
    return result

print(" : " , num_cal(num1,num2,op))
```
