# DDL(Data Definition Language)
- 테이블 같은 DB객체를 정의/수정/삭제 하는 문법


  
- 특수문자 : _ , $, # 만 가능!

  
- ※ ) PRIMARY KEY 와 UNIQUE KEY 의 차이점  : UNIQUE KEY는 NULL값이 가능 ( 중복이 없어야되지만 안적을 수도 있음  EX ) 주민등록번호 )
- FK 도 NULL 값 가능    
- CHECK : 데이터의 무결성을 유지하기 위해 테이블의 특정 컬럼에 설정

  

- 기본값 : (RESTRICT 제한) → 빌려쓰고있는 것은 삭제가 안됨  
- ON DELETE CASCADE → 삭제가 됨과 동시에, 빌려쓰고 있는 것도 같이 삭제  
- ON DELETE SET NULL → 삭제가 됨과 동시에, 빌려쓰고 있는 것은 NULL값으로 바뀜

  
- 오라클과 SQL SERVER 와의 차이점

  
- CASCADE CONSTRAINTS : 제약 조건 관계 삭제하겠다

  
- ※)DROP DELETE TRUNCATE 차이점  
- DELETE : 데이터가 모두 삭제됨 , 원상복구가 가능 , 테이블이 사라지는게 아님  
- TRUNCATE : 데이터가 모두 삭제됨, 원상복구 불가능(DDL이라서), 테이블이 사라지는게 아님  
- DROP : 데이터가 모두 삭제됨, 원상복구 불가능(DDL이라서), 테이블도 사라짐


    
- 오라클은 기본적으로 소문자를 대문자로 알아서 바꿔서 출력됨
- " " 안에는 띄어쓰기, 소문자 등 그대로 출력  
  


  
- 트랜잭션 도중에 하나라도 오류가 생기면 이전에 했던 모든 활동이 ROLLBACK 됩니다  
  
- 원자성 :ALL OR NOTHING : 트랜잭션 도중에 하나라도 오류 -> 모든 활동 롤백  
- 일관성 : PK에 중복된 값을 넣을 때 비정상 이 되면안되기때문에 제약조건을 걸어서 일관성을 유지...  
- 고립성

- 같은 트랜잭션을 동시에 실행하면 X, 한쪽 실행하고 커밋 후 실행해야함  
## AUTO COMMIT  
- 오라클 VS SQL SERVER  
  
- ※SQL SERVER에서 AUTO COMMINT = FALSE로 설정해서 문제 자주나옴  
- 오라클 : DDL -> AUTO COMMIT!! / DML -> AUTO COMMIT (X)
