```python
mask = int(input('   >> '))

if mask >=8 :
    print(f" 8  : {mask//8}")
    if mask/8 - mask//8 > 0 :
        print(f" 5  : {(mask-mask//8*8)//5+1}")
    elif mask//8 - mask/8 ==0:
        print(f" 5  : {(mask-mask//8*8)//5 }")
else :
    if mask%5 ==0:
        print(" 8  : 0")
        print(f" 5  : {mask//5}")
    else :
        print(" 8  : 0")
        print(f" 5  : {mask//5+1}")
```

```python
mask_cnt = int(input("   >> "))

#  3 
mask_8 = mask_5 =re = 0

if mask_cnt >=8 :# 8  
    mask_8 = mask_cnt//8 #8  
    mask_5 = mask_cnt%8//5 #8   
    re = mask_cnt%8%5 #8 , 5    
    if re !=0 : #  '0' 
        mask_5 += 1 #5  
    
else : # 8   
    mask_5 = mask_cnt//5
    re = mask_cnt%5
    if re !=0 : #  '0' 
        mask_5 +=1
print(f" 8  : {mask_8} \n 5  : {mask_5}")
```

```python
score = int(input("  >> "))

if 90 <= score <= 100 :
    alpha = "A"
elif 80 <= score :
    alpha = "B"
elif 70 <= score :
    alpha = "C"
elif 60 <= score :
    alpha = "D"
else :
    alpha = "F"

print(f"{score} {alpha} .")
```
