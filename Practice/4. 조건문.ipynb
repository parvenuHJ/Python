```python
num = int(input("정수 입력 >> "))
# and : 논리연산자 : True, False를 비교하는 연산자
# & : 비트연산자  : 2진수를 계산하는 연산자
# 0010 & 0111 => 0010 
if num%3==0 and num%5==0 :
    print(f"{num}은 3과 5의 배수입니다.")
else :
    print(f"{num}은 3과 5의 배수가 아닙니다.")
```


봉균이는 마스크를 대량 구매했습니다. 주변 친구들에게 소분해서 나눠주기 위해 포장지를 사려고 합니다. 포장지는 8개들이와 5개들이 2종류가 있습니다. 8개들이 포장지는 비싸기 때문에 반드시 전부 채워서 포장할 때만 사용하고 나머지는 5개들이로 포장하려고 합니다. 과연 포장지는 각각 몇 개가 필요할까요? 실행결과 처럼 동작하는 코드를 작성하세요 (단 남는 마스크 없이 전부 포장해야 합니다.)

```python
mask = int(input('마스크 개수 입력 >> '))

if mask >=8 :
    print(f"필요한 8개들이 개수 : {mask//8}개")
    if mask/8 - mask//8 > 0 :
        print(f"필요한 5개들이 개수 : {(mask-mask//8*8)//5+1}개")
    elif mask//8 - mask/8 ==0:
        print(f"필요한 5개들이 개수 : {(mask-mask//8*8)//5 }개")
else :
    if mask%5 ==0:
        print("필요한 8개들이 개수 : 0개")
        print(f"필요한 5개들이 개수 : {mask//5}")
    else :
        print("필요한 8개들이 개수 : 0개")
        print(f"필요한 5개들이 개수 : {mask//5+1}")
```

```python
mask_cnt = int(input("마스크 개수 입력 >> "))

# 변수 3개 초기화
mask_8 = mask_5 =re = 0

if mask_cnt >=8 :#마스크가 8개 이상일 때
    mask_8 = mask_cnt//8 #8로 나눈 몫
    mask_5 = mask_cnt%8//5 #8로 나눈 나머지의 몫
    re = mask_cnt%8%5 #8개들이로 담고, 5개들이로 담았을 때 남은 개수
    if re !=0 : #남은 개수가 '0'이 아니라면
        mask_5 += 1 #5개들이를 하나 더
    
else : #마스크가 8개 이상이 아닐 때
    mask_5 = mask_cnt//5
    re = mask_cnt%5
    if re !=0 : #남은 개수가 '0'이 아니라면
        mask_5 +=1
print(f"필요한 8개들이의 개수 : {mask_8}개 \n필요한 5개들이의 개수 : {mask_5}개")
```

```python
score = int(input("점수 입력 >> "))

if 90 <= score <= 100 :
    alpha = "A"
elif 80 <= score :
    alpha = "B"
elif 70 <= score :
    alpha = "C"
elif 60 <= score :
    alpha = "D"
else :
    alpha = "F"

print(f"{score}점은 {alpha}학점 입니다.")
```
