```python
# 내가 쓴 코드
score_list = [90,45,70,60,55]
n = 0

for score_list[n] in score_list:
    print(f"{n+1}번 학생은 ")
    if score_list[n] >= 60:
        print("합격입니다.")
    elif score_list[n] <60:
        print("불합격입니다.")
    n+=1
```

```python
# 선생님 코드
score_list = [90,45,70,60,55]
cnt = 0

for score in score_list :
    cnt += 1
    if score >= 60 : # 합격입니까?
        print(f'{cnt}번 학생은 합격입니다.')
    else : #불합격입니까?
        print(f'{cnt}번 학생은 불합격입니다.')
```

```python
# range () ()안에 요소들 생략 가능
# 시작값 생략가능 -> 생략하면 "0" 이 기본값
for i in range(10) :
    print(i)
```

```python
# range () ()안에 요소들 생략 가능
# 증가량 생략가능 -> 생략하면 "1" 이 기본값
for i in range(1,10) :
    print(i)
```

```python
# 선생님 답
user_input = 'a'
menu = ['햄버거','피자','음료수']
order = []


while len(order) < len(menu):
    cnt = 0
    print("=== 키오스크 ===")
    for i in menu:
        print(f"{cnt+1}, {i}")
        cnt +=1
    print()
    user_input = input("메뉴를 선택하거나 'q'를 눌러 종료하세요:  ")
    
    if user_input == 'q':
        break
    elif user_input not in ['1','2','3'] :
        print()
        print("올바른 메뉴 번호를 선택해주세요")
        print("-"*30)
        continue  #다시 while문으로 올라감 #여기서는 안써도 똑같지만 이런 기능
                   # 나중에 잘 써먹자
    else :
        order.append(menu[int(user_input)-1])
        print()
        print(f'{menu[int(user_input)-1]}를 선택하셨습니다.\n')
        print('-'*30)

print('=== 주문한 메뉴 ===')
for item in order:
    print(item)
    
        
```
