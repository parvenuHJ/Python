```python
#  
user_input = 'a'
menu = ['','','']
order = []


while len(order) < len(menu):
    cnt = 0
    print("===  ===")
    for i in menu:
        print(f"{cnt+1}, {i}")
        cnt +=1
    print()
    user_input = input("  'q'  :  ")
    
    if user_input == 'q':
        break
    elif user_input not in ['1','2','3'] :
        print()
        print("   ")
        print("-"*30)
        continue  # while  #    
                   #   
    else :
        order.append(menu[int(user_input)-1])
        print()
        print(f'{menu[int(user_input)-1]} .\n')
        print('-'*30)

print('===   ===')
for item in order:
    print(item)
    
        
```
